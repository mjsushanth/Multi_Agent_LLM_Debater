## Quick Check On Available Models.

````python
import requests
import time
import subprocess


def check_ollama_status():
    """Check basic Ollama API status"""
    try:
        response = requests.get("http://localhost:11434/api/tags", timeout=5)
        print(f"API Status: {response.status_code}")
        if response.status_code == 200:
            models = response.json().get("models", [])
            print(f"Available models: {[m['name'] for m in models]}")
        return response.status_code == 200
    except Exception as e:
        print(f"Error connecting to Ollama: {e}")
        return False

check_ollama_status()
````

````python
import requests

def check_model_sizes():
    try:
        response = requests.get("http://localhost:11434/api/tags")
        
        if response.status_code != 200:
            print(f"Error: {response.status_code}")
            return
            
        models = response.json().get("models", [])
        
        print(f"{'Model Name':<30} {'Size (GB)':<10} {'Modified Date':<20}")
        print("-" * 60)
        
        total_size = 0
        for model in models:
            name = model.get("name", "Unknown")
            size_bytes = model.get("size", 0)
            size_gb = round(size_bytes / (1024 * 1024 * 1024), 2)
            modified = model.get("modified", "Unknown")
            
            total_size += size_bytes
            print(f"{name:<30} {size_gb:<10.2f} {modified:<20}")
            
        total_size_gb = round(total_size / (1024 * 1024 * 1024), 2)
        print("-" * 60)
        print(f"Total size: {total_size_gb} GB")
        
    except Exception as e:
        print(f"Error: {e}")

check_model_sizes()
````

## Code for deleting a model:

````python
import subprocess

def remove_ollama_model(model_name):
    """Remove an Ollama model"""
    try:
        result = subprocess.run(
            ["ollama", "rm", model_name],
            capture_output=True,
            text=True,
            check=True
        )
        print(f"Successfully removed model: {model_name}")
        print(result.stdout)
        return True
    except subprocess.CalledProcessError as e:
        print(f"Error removing model: {e}")
        print(e.stderr)
        return False


# remove_ollama_model("deepseek-r1:7b")

# remove_ollama_model("deepseek-llm:7b")

# remove_ollama_model("llama3.1:8b")

remove_ollama_model("yi:9b")

````

```
Windows filesystem - Where your models are stored on the E: drive
VM boundary - The hypervisor layer between Windows and the Linux kernel
WSL2 filesystem - The ext4 filesystem inside the VM

When Ollama (running in Windows) loads a model from your E: drive, it's accessing files directly. But when you run it through WSL2, several layers of translation occur.

When you access Windows drives from WSL2, the following happens:
    The request starts in the Linux environment
    It passes through a filesystem driver called 9P (Plan 9) protocol
    This crosses the VM boundary to Windows
    Windows handles the actual file access to your E: drive
    Data is then passed back across the VM boundary
    It returns through 9P to your Linux application
    
Each large read operation requires:
    Serialization/deserialization of requests
    Context switching between the VM and host
    Memory copies across protection boundaries
    Protocol overhead from the 9P filesystem

Native Windows access: ~1-2 GB/s read speed
WSL2 accessing Windows drives: ~200-400 MB/s read speed (5x slower)
Large file operations: Up to 20x slowdown in some cases

```
